```python
from mwviews.api import PageviewsClient
import pandas as pd
import matplotlib. pyplot as plt
p = PageviewsClient(user_agent="kez070@ucsd.edu Selfie, Cat, and Dog analysis")
```

```python
articles =  ["Albedo", "Talk:Atlas_Shrugged", "Alabama", "Achilles", "Ayn_Rand", "Altruism", "Andre_Agassi"]

values = p.article_views('en.wikipedia',articles, granularity='monthly', start='20150901', end='20191101')
```

```python
print(values)
```

```python
for i in values.keys():
    print(values[i])
```

```python
all_keys = list(values.keys())
```

```python
all_keys.sort()
```

```python

val_dict = []
for x in articles:
    print(x)
    for key in all_keys:
        val_dict.append({"name":x,"time":key, "views":values[key][x]})
```

```python
df = pd.DataFrame(val_dict)
df.groupby('name')
```

```python

df =df.fillna(0)
```

```python
fig, ax = plt.subplots(figsize=(15,7))
df.groupby(['time','name']).sum()['views'].unstack().plot(ax=ax)
```

```python
articles = open('article_titles.txt', 'r') 
article_names = []
Lines = articles.readlines() 
for line in Lines:
    article_names.append(line.rstrip())
```

```python
main_dfs = []
for i in range(100,len(article_names),100):
    values = p.article_views('en.wikipedia',article_names[i-100:i], granularity='monthly', start='20020901', end='20191101')
    all_keys = list(values.keys())
    all_keys.sort()
    val_dict = []
    for x in article_names[i-100:i]:
        for key in all_keys:
            val_dict.append({"name":x,"time":key, "views":values[key][x]})
    df = pd.DataFrame(val_dict)
    main_dfs.append(df)
    print(i)
```

```python
main_dfs
```

```python
for i in range(len(main_dfs)):
    main_dfs[i] = main_dfs[i].fillna(0)
```

```python
df = pd.concat(main_dfs)
```

```python
df.to_csv('page_views.csv')
```

```python
fig, ax = plt.subplots(figsize=(15,7))
main_dfs[0].groupby(['time','name']).sum()['views'].unstack().plot(ax=ax)
```
